```python
filtered_words =[word for word in movie_reviews.words()
                if not word in useless_words]
type(filtered_words)
```

```python
from collections import Counter
word_counter = Counter(filtered_words)
```

```python
most_common_words =word_counter.most_common() [:10]
```

```python
sorted_word_counts = sorted(list (word_counter.values()), reverse =True)
plt.loglog(sorted_word_counts)
plt.ylabel("Freq")
plt.xlabel("Word Rank")
```

```python
negative_features =[
(build_bag_of_words_features_filtered
(movie_reviews.words(fileids =[f])), 'neg')\
 for f in negative_fileids]
```

```python
print(negative_features [6])
```

```python
positive_features =[
(build_bag_of_words_features_filtered
(movie_reviews.words(fileids =[f])), 'pos')\
 for f in positive_fileids]
```

```python
print(positive_features [6])
```

```python
from nltk.classify import NaiveBayesClassifier
```

```python
split =700
```

```python
sentiment_classifier =NaiveBayesClassifier.train(positive_features[:split]+negative_features[:split] )
```

```python
nltk.classify.util.accuracy
(sentiment_classifier,positive_features[:split]+negative_features[:split]) *100
```

```python
sentiment_classifier.show_most_informative_features()
```
